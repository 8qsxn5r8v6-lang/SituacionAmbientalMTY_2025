##Limpieza de los datos

```python
import pandas as pd

# Lista de columnas que quieres conservar
columnas_a_conservar = [
    'datetime','CENTRO_CO', 'CENTRO_NOX', 'CENTRO_PM10', 'CENTRO_PM2.5', 'CENTRO_O3',
    'NORESTE2_CO', 'NORESTE2_NOX', 'NORESTE2_O3', 'NORESTE2_PM10', 'NORESTE2_PM2.5',
    'NORESTE_CO', 'NORESTE_NOX', 'NORESTE_O3', 'NORESTE_PM10', 'NORESTE_PM2.5',
    'SURESTE2_CO', 'SURESTE2_NOX', 'SURESTE2_O3', 'SURESTE2_PM10', 'SURESTE2_PM2.5',
    'SUR_CO', 'SUR_NOX', 'SUR_O3', 'SUR_PM10', 'SUR_PM2.5'
]

# Cargar el archivo Excel
df = pd.read_excel('/content/master_unificado_mapped_wide (2).xlsx')

# Verificar qué columnas existen en el dataframe
columnas_existentes = [col for col in columnas_a_conservar if col in df.columns]
columnas_faltantes = [col for col in columnas_a_conservar if col not in df.columns]

print(f"Columnas encontradas: {len(columnas_existentes)}")
print(f"Columnas faltantes: {columnas_faltantes}")

# Crear nuevo dataframe solo con las columnas deseadas
df_limpio = df[columnas_existentes]

# Guardar el nuevo archivo limpio
df_limpio.to_excel('master_limpio.xlsx', index=False)

print("¡Limpieza completada! Archivo guardado como 'master_limpio.xlsx'")
df_limpio.head(10)
```

##Exploracion estadistica de los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Configuración para mejores visualizaciones
plt.style.use('default')
sns.set_palette("husl")

def analisis_estadistico_completo(df):
    print("=" * 60)
    print("ANÁLISIS ESTADÍSTICO COMPLETO")
    print("=" * 60)

    # Identificar tipos de variables
    variables_cuantitativas = df.select_dtypes(include=[np.number]).columns.tolist()
    variables_cualitativas = df.select_dtypes(include=['object', 'category']).columns.tolist()

    print(f"\nVariables cuantitativas encontradas: {len(variables_cuantitativas)}")
    print(f"Variables cualitativas encontradas: {len(variables_cualitativas)}")

    # ANÁLISIS PARA VARIABLES CUANTITATIVAS
    if variables_cuantitativas:
        print("\n" + "=" * 50)
        print("ANÁLISIS DE VARIABLES CUANTITATIVAS")
        print("=" * 50)

        # Dataframe para almacenar resultados
        resumen_cuantitativas = pd.DataFrame(index=variables_cuantitativas)

        for variable in variables_cuantitativas:
            print(f"\n--- {variable} ---")
            datos = df[variable].dropna()

            # Medidas de tendencia central
            media = datos.mean()
            mediana = datos.median()
            moda = datos.mode()
            moda_valor = moda.iloc[0] if not moda.empty else np.nan

            # Medidas de dispersión
            rango = datos.max() - datos.min()
            varianza = datos.var()
            desviacion_std = datos.std()
            q1 = datos.quantile(0.25)
            q3 = datos.quantile(0.75)
            iqr = q3 - q1

            # Almacenar en el resumen
            resumen_cuantitativas.loc[variable, 'Media'] = media
            resumen_cuantitativas.loc[variable, 'Mediana'] = mediana
            resumen_cuantitativas.loc[variable, 'Moda'] = moda_valor
            resumen_cuantitativas.loc[variable, 'Mínimo'] = datos.min()
            resumen_cuantitativas.loc[variable, 'Máximo'] = datos.max()
            resumen_cuantitativas.loc[variable, 'Rango'] = rango
            resumen_cuantitativas.loc[variable, 'Varianza'] = varianza
            resumen_cuantitativas.loc[variable, 'Desviación_Std'] = desviacion_std
            resumen_cuantitativas.loc[variable, 'Q1'] = q1
            resumen_cuantitativas.loc[variable, 'Q3'] = q3
            resumen_cuantitativas.loc[variable, 'IQR'] = iqr

            print(f"Media: {media:.2f}")
            print(f"Mediana: {mediana:.2f}")
            print(f"Moda: {moda_valor:.2f}")
            print(f"Rango: {rango:.2f}")
            print(f"Varianza: {varianza:.2f}")
            print(f"Desviación estándar: {desviacion_std:.2f}")
            print(f"Q1: {q1:.2f}, Q3: {q3:.2f}, IQR: {iqr:.2f}")

        # VISUALIZACIONES PARA VARIABLES CUANTITATIVAS
        print("\n" + "=" * 40)
        print("GENERANDO VISUALIZACIONES CUANTITATIVAS")
        print("=" * 40)

        # Boxplots para detectar outliers
        if len(variables_cuantitativas) > 0:
            fig, axes = plt.subplots(2, 2, figsize=(15, 12))

            # Boxplots
            df[variables_cuantitativas].boxplot(ax=axes[0,0], rot=45)
            axes[0,0].set_title('Boxplots - Detección de Outliers')
            axes[0,0].tick_params(axis='x', rotation=45)

            # Histogramas
            if len(variables_cuantitativas) >= 4:
                for i, var in enumerate(variables_cuantitativas[:4]):
                    row, col = i//2, i%2
                    axes[1, col].hist(df[var].dropna(), bins=20, alpha=0.7, edgecolor='black')
                    axes[1, col].set_title(f'Histograma de {var}')
                    axes[1, col].set_xlabel(var)
                    axes[1, col].set_ylabel('Frecuencia')
            else:
                for i, var in enumerate(variables_cuantitativas):
                    axes[1, i].hist(df[var].dropna(), bins=20, alpha=0.7, edgecolor='black')
                    axes[1, i].set_title(f'Histograma de {var}')
                    axes[1, i].set_xlabel(var)
                    axes[1, i].set_ylabel('Frecuencia')

            plt.tight_layout()
            plt.savefig('visualizaciones_cuantitativas.png', dpi=300, bbox_inches='tight')
            plt.show()

        # Mapa de calor de correlaciones
        if len(variables_cuantitativas) > 1:
            plt.figure(figsize=(12, 8))
            correlacion = df[variables_cuantitativas].corr()
            mask = np.triu(np.ones_like(correlacion, dtype=bool))
            sns.heatmap(correlacion, mask=mask, annot=True, cmap='coolwarm', center=0,
                       square=True, fmt='.2f')
            plt.title('Mapa de Calor - Correlación entre Variables')
            plt.tight_layout()
            plt.savefig('mapa_calor_correlacion.png', dpi=300, bbox_inches='tight')
            plt.show()

            print("\nMatriz de Correlación:")
            print(correlacion.round(2))

    # ANÁLISIS PARA VARIABLES CUALITATIVAS
    if variables_cualitativas:
        print("\n" + "=" * 50)
        print("ANÁLISIS DE VARIABLES CUALITATIVAS")
        print("=" * 50)

        for variable in variables_cualitativas:
            print(f"\n--- {variable} ---")
            datos = df[variable].dropna()

            # Tabla de distribución de frecuencia
            freq_absoluta = datos.value_counts()
            freq_relativa = datos.value_counts(normalize=True) * 100

            tabla_frecuencias = pd.DataFrame({
                'Frecuencia Absoluta': freq_absoluta,
                'Frecuencia Relativa (%)': freq_relativa.round(2),
                'Frecuencia Acumulada': freq_absoluta.cumsum()
            })

            print("Tabla de distribución de frecuencia:")
            print(tabla_frecuencias)

            # VISUALIZACIONES PARA VARIABLES CUALITATIVAS
            if len(freq_absoluta) > 0:
                fig, axes = plt.subplots(1, 2, figsize=(15, 6))

                # Diagrama de barras
                freq_absoluta.plot(kind='bar', ax=axes[0], color='skyblue', edgecolor='black')
                axes[0].set_title(f'Distribución de {variable}')
                axes[0].set_xlabel(variable)
                axes[0].set_ylabel('Frecuencia')
                axes[0].tick_params(axis='x', rotation=45)

                # Diagrama de pastel (solo si no hay demasiadas categorías)
                if len(freq_absoluta) <= 10:
                    axes[1].pie(freq_relativa, labels=freq_absoluta.index, autopct='%1.1f%%', startangle=90)
                    axes[1].set_title(f'Distribución de {variable} (%)')
                else:
                    axes[1].text(0.5, 0.5, 'Demasiadas categorías\npara diagrama de pastel',
                               ha='center', va='center', transform=axes[1].transAxes)
                    axes[1].set_title(f'Distribución de {variable}')

                plt.tight_layout()
                plt.savefig(f'distribucion_{variable}.png', dpi=300, bbox_inches='tight')
                plt.show()

    # RESUMEN GENERAL
    print("\n" + "=" * 60)
    print("RESUMEN GENERAL DEL DATASET")
    print("=" * 60)
    print(f"Total de registros: {len(df)}")
    print(f"Total de variables: {len(df.columns)}")
    print(f"Variables cuantitativas: {len(variables_cuantitativas)}")
    print(f"Variables cualitativas: {len(variables_cualitativas)}")

    if variables_cuantitativas:
        print("\nResumen estadístico de variables cuantitativas:")
        print(resumen_cuantitativas.round(2))

        # Guardar resumen en Excel
        resumen_cuantitativas.to_excel('resumen_estadistico_cuantitativas.xlsx')
        print("\nResumen guardado en 'resumen_estadistico_cuantitativas.xlsx'")

    return resumen_cuantitativas if variables_cuantitativas else None

# CARGAR TU ARCHIVO LIMPIO
print("Cargando archivo limpio...")
df = pd.read_excel('/content/master_limpio.xlsx')

# EJECUTAR ANÁLISIS COMPLETO
resultados = analisis_estadistico_completo(df)

print("\n¡Análisis completado! Se han generado:")
print("✓ Gráficos de visualización")
print("✓ Resúmenes estadísticos")
print("✓ Archivos Excel con resultados")
```
